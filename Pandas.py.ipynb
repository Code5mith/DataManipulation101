# Full Pandas guide First import the library

```python
# You can install the library by using pip install pandas
import pandas as pa
```

```python
# Reading data from a csv or any supported file 

data = pa.read_csv("student-mat.csv", delimiter=";")

#Pandas returnes the data from the file as a dataframe which is like a matrix or a table .

data
```

# To access data from the dataframe

```python
# Pandas uses the header of the file as an index to each column so we can access each column with its header as an 
# index this property gives you all the headers or the first row .

data.columns

# How to use the header to access the column 

print(data['age'])

# You can also pass a list of headers to access multiple columns

headers = ['school', 'age', 'address']

print('\n', data[headers])

# To get a specific row we can use the iloc() method on our dataframe
print(data.iloc[0])

# we can also loop through all the rows in our dataframe by the iterrow() function 
# Since iterrow() returns the index of the  row and the row itself we have to use to variables 
 
print("\n")
for index, item in data.iterrows():
    print(index, item['school'])

```

# To get a full description of the whole data 

```python
# We can also get a full summery of the dataframe with one method describe().

data.describe()
```

## we can also sort values based lettering or number or any other way

```python
# You can also reorganise the dataframe based on a specific row either in ascending or descending order

data.sort_values('age', ascending=False)

#sorts the dataframe with decreasing age of each item
```

# Making changes to the dataframe

```python
# To add a new column in the dataframe we can either use existing data to set the new data or import a new data 
# to the new column

# Here we are setting the new row to be the sum 'G1' 'G2' 'G3' columns.

data['totalGrades'] = data['G1'] + data['G2'] + data['G3']

data.head()
```

# Droping a column form the dataframe

```python
# We can also remove any columns that we dont want in the dataframe by using the drop() method.

data['totalGrades'] = data['G1'] + data['G2'] + data['G3']
print(data)


data = data.drop(columns=['totalGrades'])

data
```

# Advanced data filtering

```python

print(data[(data['G1'] > 10) & (data['G2'] > 18)])

new_data = data[(data['G1'] > 10) & (data['G2'] > 15)]


print('\n \n++++++++++++++++++++++++++++++++++++++++++++++++++++++ \n \n')

# Reseting the index of each row

new_data = new_data.reset_index()

print(new_data)
```

```python
# Changing values in our dataframe

data.loc[data['G1'] == 0 , 'G1'] = 10110001

data
```

```python
# Performing different mathimatical operations on our dataframe

# Grouping different attributes by there mean or average

print('By mean', '\n', data.groupby(['school']).mean())

#Grouping different attributes by there sum

print('\n', 'By sum', '\n', data.groupby(['school']).sum())

# Grouping by counting the number of items in each column 
print('\n', data.groupby(['school', 'sex', 'address', 'famsize']).count().head())
```
