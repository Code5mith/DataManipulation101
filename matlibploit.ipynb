# Necessary libraries 

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Basic Graph dev

#### Every function in  matplotlib is  available on the documentation page 

```python
#To plot a basic graph we need two 1D arrays to be passed to the plot function in plt 

x = [2,3,4,5,6]
y = [5,6,7,8,9]

plt.plot(x, y)

#to add title and fontdict is to specify the font properties we can include it in any of the functions below 
plt.title("first graph", fontdict={"fontname": "quicksand", "fontsize": 25})

#for lablels
plt.xlabel("x-axis", fontdict={"fontname": "quicksand", "fontsize": 18})
plt.ylabel("y-axis", fontdict={"fontname": "quicksand", "fontsize": 18})
```

```python
#the x and y values are not the same as the once in out array so to do that we can use
plt.xticks(x)
plt.yticks(y)

plt.plot(x, y)
```

```python
#To add a discripton on each plot in our graph we use the legend() function 
plt.xticks(x)
plt.yticks(y)

#plot should come first
plt.plot(x, y, label="describe the graph or line")

plt.legend()

plt.show()
```

```python
#to customise the graph we can pass the properties to the plot function 
plt.plot(x,y, color="red", linewidth="2", marker=".", markersize='9', markeredgecolor="blue", linestyle="dotted")

plt.show()
```

```python
#there is a short hand method to pass the properties to the plot func

#format is [color][marker][linestyle] (r.-- in the code) same as above
plt.plot(x,y, 'r.--', markersize='9', markeredgecolor="blue", label='dotted graph')

plt.legend()

plt.show()
```

```python
# Plotting multiple graphs 

# arange() to initialise our arrays 

a = np.arange(2,8,0.5)

print(a, "\n")

plt.plot(x,y, 'r.--', markersize='9', markeredgecolor="blue", label="firstgraph")

#plotting for the first 6 items in a 

plt.plot(a[:6], a[:6]**2, 'b', marker=".", label="square of a")

# How to show projections

plt.plot(a[5:], a[5:]**2, 'orange', marker=".", linestyle="dashed", label="future projection")

plt.legend()

plt.show()
```

```python
# How to resize our plots

plt.figure(figsize=(5,3), dpi=100)

plt.plot(x,y)
```

```python
# Exporting our graphs to a image file

#make sure to plot it first

plt.plot(x,y, 'r.--', markersize='9', markeredgecolor="blue", label="firstgraph")

plt.savefig("firstgraph.png", dpi=400)
```

# Bar charts

```python

a = np.arange(2,7)
b = np.arange(3,8)


# Change the plot() to bar()  

plt.bar(a,b, color="red")

plt.xticks(a)
plt.yticks(b)

#additional functionalities you can access each bar with index as an array

bars = plt.bar(a,b)

#to hatch the bar
bars[0].set_hatch('/')
bars[2].set_hatch('*')

#To change the color of a single bar
plt.bar(a[3],b[3], color="red")
```
