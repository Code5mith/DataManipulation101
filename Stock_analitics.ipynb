```python
## Filter the data by time 

months = de(['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec'])

filtered_dict = {}


def _json_filter(year):
    for x in range(12):
        month = x + 1
        filtered_dict[months[x]] = de([])
        for y in range(1,32):
            day = y + 1
            _temp = df2[(df2['Date'] >= f"{year}-{f'0{month}' if x < 9 else month}-{f'0{day}' if y < 9 else day}") 
                        & (df2['Date'] < f"{year}-{f'0{month}' if x < 9 else month}-{f'0{day + 1}' if y < 8 else day + 1}")][header]
            filtered_dict[months[x]].append(_temp)
            
            
            #print(_temp.loc(_temp.index)[:])
            
def null_filter(dicto):
    for x in range(12):
        for y in range(31):
            if 'yes' if len(dicto[months[x]][y]) == 0 else 'no' == 'yes':
                dicto[months[x]][y] = 0
            else:
                continue
                
_json_filter(2016)

null_filter(filtered_dict)

filtered_dict

```

```python
# extract min and max values for each month

monthly_mean = {}

def monthly_mean_func(year):
    
    for x in range(12):
        monthly_mean[months[x]] = de([])
        month = x + 1
        
        for y in header[:-1]:
            
            _temp = df2[(df["Date"] >= f"{year}-{f'0{month}' if x < 9 else month}") 
                            & (df["Date"] < f"{year}-{f'0{month + 1}' if x < 8 else month + 1}")][y].mean()
            monthly_mean[months[x]].append(_temp)
           

        
monthly_mean_func(2015)

monthly_mean
```

```python
#plot the monthly price action for a year

plt.figure(figsize=(15,10), dpi=100, layout='constrained')

#sas.plot(monthly_mean)

for x,y in monthly_mean.items():
    
    plt.bar(x,sum(y)/len(y), label=x)

plt.title('Monthly average price action')
plt.legend(loc='upper right')
plt.show()
```

```python
# sample pi charts for each month average price
def piplot(month):
    
    plt.pie(monthly_mean[month], labels=[f"{header[0]} -> {monthly_mean[month][0]}",
                                         f"{header[1]} -> {monthly_mean[month][1]}",
                                         f"{header[2]} -> {monthly_mean[month][2]}",
                                         f"{header[3]} -> {monthly_mean[month][3]}"])
    
piplot('may')
```
