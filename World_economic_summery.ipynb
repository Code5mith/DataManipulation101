```python
import matplotlib.pyplot as plt
import math
import sys
import time
import numpy as np
import pandas as pa
from collections import deque as de

```

```python
Dset1 = pa.read_csv('World_economy.csv')

headers = Dset1.columns[4:-1]

World_economy = np.array(Dset1.iloc[259][4:-1])

#helper function for plotting
def plotter(x_axis,y_axis,marker,markeredgecolor,linewidth,lable,lib=plt):
    
    out=lib.plot(x_axis,y_axis,marker=marker,label=lable,markeredgecolor=markeredgecolor,linewidth=linewidth)
    return out
def checker_first(lists):
    for x in range(len(lists)):
        if np.isnan(lists[x]):
            lists[x] = 0.0
        else:
            continue
checker_first(World_economy)


#a.empty, a.bool(), a.item(), a.any() or a.all()
```

```python

# United Arab Emirates 	ARE  China 	CHN  Europe & Central Asia 	ECS   Euro area 	EMU    European Union 	EUU  
# United Kingdom 	GBR 	  GAB  Japan 	JPN  Russian Federation 	RUS  	United States 	USA  
```

```python
# Plotting the world economy graph 

plt.figure(figsize=(15,10), dpi=100, layout='constrained')

plt.title('World Economy 2000-21')

plotter(headers[40:-1], World_economy[40:-1], "x" , "red", 1,"GDP in Billions")

plt.yticks(list( World_economy[40:-1]))

plt.legend()

plt.savefig('World Economy.png', dpi=300)

plt.show()
```

```python
# Plotting GDP of top 5 countries.

Top_6 = Dset1.iloc[[8,251,35,55,81,119]]

Top_6 = Top_6.drop(columns='Unnamed: 67')

Top_6 = Top_6.reset_index()

country_name = ['UnitedArabEmirates','UnitedStates','Canada','Germany','UnitedKingdom','Japan']

UnitedArabEmirates = np.array(Top_6[0:1][headers])
UnitedStates = np.array(Top_6[1:2][headers])
Canada = np.array(Top_6[2:3][headers])
Germany = np.array(Top_6[3:4][headers])
UnitedKingdom = np.array(Top_6[4:5][headers])
Japan = np.array(Top_6[5:6][headers])

Grand_total = de([])

def checker(lists):
    for x in range(len(lists[0])):
        if np.isnan(lists)[0][x]:
            lists[0][x] = 0
        else:
            continue
    Grand_total.append(lists.sum())

checker(UnitedArabEmirates )
checker(UnitedStates)
checker(Canada)
checker(Germany)
checker(UnitedKingdom)
checker(Japan)

Country_dict = {}

i = 0
for x in country_name:
    Country_dict[x] = Grand_total[i]
    i+=1
    
Country_dict

```

```python
# Plotting GDP of top 5 count

plt.figure(figsize=(15,5), dpi=100, layout='constrained')

y = list(Grand_total)
x = list(country_name)

plt.plot(x,y,marker='o',label='Total GDP')

plt.title('Top 5 countries total gpd 1960-2021',  fontdict={"fontname": "serif", "fontsize": 18})

plt.yticks(Grand_total)

plt.xlabel('Country Name', fontdict={"fontname": "serif", "fontsize": 18})
plt.ylabel('Total GDP',  fontdict={"fontname": "serif", "fontsize": 18})

plt.legend()

plt.show()
```

```python
# GPD for every country in the top five list

# Pi chart

labels = country_name
sizes = Grand_total

plt.pie(sizes, labels=labels)
plt.title('Wealth comparision of the top  five countries from 1960-2021')

plt.show()
```
